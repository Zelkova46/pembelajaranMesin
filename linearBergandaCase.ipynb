```python
#import library
import numpy as np
import math 
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
dataset = pd.read_csv('pengeluaranKeluarga.csv')
x2 = dataset.iloc[:, 2].values
x1 = dataset.iloc[:, 1].values
y = dataset.iloc[:, 0].values
```

```python
print(x1)
print(x2)
print(y)
```

```python
#mengetahui jumlah data observasi
n = len(y)
print("Jumlah data yang (n) : ",n)
```

## mengumpulkan nilai hasil olah data observasi

```python
sigmaY = sum(y)
print(sigmaY)
```

```python
sigmaX1 = sum(x1)
print(sigmaX1)
```

```python
sigmaX2 = sum(x2)
print(sigmaX2)
```

```python
x1kuadrat = pow(x1, 2)
sigmax1kuadrat = sum(x1kuadrat)
print(sigmax1kuadrat)
```

```python
x2kuadrat = pow(x2, 2)
sigmax2kuadrat = sum(x2kuadrat)
print(sigmax2kuadrat)
```

```python
x1x2 = x1 * x2
sigmax1x2 = sum(x1x2)
print(sigmax1x2)
```

```python
x1y = x1 * y
sigmax1y = sum(x1y)
print(sigmax1y)
```

```python
x2y = x2 * y
sigmax2y = sum(x2y)
print(sigmax2y)
```

## memasukkan data yang telah diolah tadi ke dalam bentuk matriks

```python
A = [[n, sigmaX1, sigmaX2], 
    [sigmaX1, sigmax1kuadrat, sigmax1x2],
    [sigmaX2, sigmax1x2, sigmax2kuadrat]]
print(A)
```

```python
H = [sigmaY, sigmax1y, sigmax2y]
print(H)
```

## menghitung rumus b = A invers * H

```python
#menentukan invers dari matriks A menggunakan numpy
Ainvers = (np.linalg.inv(A))
print(Ainvers)
```

```python
# menghitung elemen elemen yang ada pada matriks b menggunakan numpy

b = np.dot(Ainvers,H)
print(b)
```

```python
print("Elemen dari matriks b menghasilkan model regresi linier berganda :")
persRegresiGanda = " Y = {b0} + {b1} X1 + {b2} X2".format(b0 = b[0], b1 = b[1], b2 = b[2])
print(persRegresiGanda)
```
