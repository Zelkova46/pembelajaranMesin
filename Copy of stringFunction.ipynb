```python
import nltk
import numpy as np
import pandas as pd
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
text = "This's a sent tokenize test. this is sent two. is this sent three? sent 4 is cool!  Now it's your turn."
```

```python
# KALO MAU NGEBIKIN STRING JADI LOWER
textLower = text.lower()
```

```python
import nltk
nltk.download('punkt')
```

Artikel text yang ada di variabel text dipecah per kalimat.

```python
kal = textLower
kalimat = nltk.sent_tokenize(kal)
print(kalimat)
```

Pecah text menjadi per kalimat

```python
#cara 1
words = nltk.word_tokenize(textLower)
print(words)
```

```python
#cara 2
words2 = [word.lower() for word in words if word.isalpha()]
print(words2)
```

Perbedaan dari cara 1 dan cara dua adalah dengan menggunakan library dari nltk dapat mengambil token perkata dari kalimat yang diberikan dan juga dapat mengambil tanda bacanya juga.

Sementara untuk cara memenggal kata yang kedua adalah dengan memanfaatkan for loop, kata kata di string di perkecil dahulu lalu pemenggalan dilakukan dengan melakukan for loop, jika ada karakter yang bukan alphabet didalam string maka kata sebelumnya putus dan ditampilkan, maka dari itu pada cara yang kedua tidak dapat menampilkan tanda baca

```python
jmlktwr = Counter(words).most_common()
```

```python
words_low = text.lower()
print(words_low)
```

```python
print(jmlktwr)
```

```python
print(type(jmlktwr))
```

Hitunga bobot kata berdasarkan rata-rata. Rumus = jumlah kata tsb / total jumlah kata dalam 1 kalimat

Susun kata dan bobotnya kedalam lIst

```python
lsgabok = [ ]
lskt = []
lsbbt = []

tkata = len(words)
for kt, jkt in jmlktwr:
  jbbt = jkt/tkata
  lskt.append(kt)
  lsbbt.append(jbbt)
  lsgabok.append((kt, jbbt))
  print(kt, jkt, jbbt)
```

## Cara lain dengan textBlob

```python
K1 = "Hari ini saya senang sekali"
K2 = "Besok kita akan pergi ke luar kota"
K3 = "Hari yang cerah dan menyenangkan untuk pergi ke pasar"
```

```python
bloblist = []
```

```python
from textblob import TextBlob as tb
```

```python

bloblist.append(tb(K1))
bloblist.append(tb(K2))
bloblist.append(tb(K3))
```

```python
bloblist
```

```python
lisall = []
for i, blob in enumerate(bloblist):
  print('Kalimat ke : ', i)
  print(blob)
  print(blob.words.lower())
  llst = blob.words.lower()
  lisall.append(llst)
  
```

```python
print(lisall)
```
